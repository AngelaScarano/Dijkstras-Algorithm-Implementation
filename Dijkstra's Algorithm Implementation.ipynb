```python
#Supporting functions and constants
import collections
import math
import sys

GRAPH_X_SIZE = 10
GRAPH_Y_SIZE = 10

#arbitrary, very large maximum value
MAX_INT = 999999999999999999

theMap =[[60, 15, 3, 4, 5, 6, 7, 8, 9, 15], 
         [4, 0, 8, 4, 38, 5, 3, 11, 2, 4], 
         [5, 10, 20, 2, 8, 4, 9, 9, 2, 8], 
         [4, 4, 30, 0, 45, 14, 2, 7, 9, 6], 
         [5, 4, 6, 2, 9, 10, 1, 2, 2, 3], 
         [4, 4, 0, 14, 10, 6, 2, 7, 3, 5], 
         [5, 4, 0, 5, 3, 2, 9, 1, 6, 1], 
         [8, 11, 7, 7, 7, 0, 1, 8, 7, 4], 
         [4, 4, 2, 11, 8, 0, 15, 0, 0, 30],
         [5, 4, 2, 9, 9, 9, 6, 7, 40, 50]
         ]; 

class Node:
    def __init__(self, data, xPos, yPos):
        self.parent = None
        self.child = None
        self.data = data
        self.xPos = xPos
        self.yPos = yPos
        self.distance = MAX_INT
        
def terrain (map, algorithm):
    if algorithm == "D":
        Dijkstra(map)
```

```python
#Function to implement Dijkstra's single source shortest path algorithm for a graph with ajacency matrix representation

def Dijkstra(map):
    #create an unexplored set of nodes for each position in the graph terrain
    unExSet = []
    for i in range(0, (GRAPH_Y_SIZE)):
        for j in range(0, (GRAPH_X_SIZE)):
            newNode = Node(map[i][j], j, i)
            unExSet.append(newNode)
            
    #Checked that all list items were the correct nodes
    #listLen = len(unExSet)
    #for i in range(0, listLen):
            #print(unExSet[i].data, unExSet[i].xPos, unExSet[i].yPos)
    
    #start at root node, remove root from unExSet
    sptSet = collections.deque([])
    rootNode = unExSet[0]
    rootNode.distance = rootNode.data
    unExSet.remove(unExSet[0])
    sptSet.append(rootNode)
    currNode = rootNode
    finalNode = None
    
    mazeEnd = False
    #continue checking for a path while the unExSet isn't empty
    #while unExSet is not empty:
    while mazeEnd != True:
        minVal = MAX_INT
        listLen = len(unExSet)
        tempNode = currNode
        nodeFound = False
        
        #check if down position exists and is best
        if (currNode.yPos - 1) > 0:
            #check if node exists in unExSet. If it exists, update its distance value and check if it's the lowest value move
            for i in range(0, (listLen)):
                if unExSet[i].yPos == (currNode.yPos - 1) and unExSet[i].xPos == (currNode.xPos):
                    nodeFound = True
                    unExSet[i].distance = unExSet[i].data
                    if unExSet[i].distance < minVal:
                        minVal = unExSet[i].distance
                        tempNode = unExSet[i]
                    
                if nodeFound == True:
                    break
        
        listLen = len(unExSet)
        nodeFound = False
        #check if the right position exists and is best
        if (currNode.xPos + 1) < GRAPH_X_SIZE:
            #check if node exists in unExSet. If it exists, update its distance value and check if it's the lowest value move
            for i in range(0, (listLen)):
                if unExSet[i].xPos == (currNode.xPos + 1) and unExSet[i].yPos == (currNode.yPos):
                    nodeFound = True
                    unExSet[i].distance = unExSet[i].data
                    
                    #prioritizes moving into the final sqare
                    if unExSet[i].xPos == GRAPH_X_SIZE-1 and unExSet[i].yPos == GRAPH_Y_SIZE:
                        minVal = -1
                        tempNode = unExSet[i]
                        
                    if unExSet[i].distance < minVal:
                        minVal = unExSet[i].distance
                        tempNode = unExSet[i]
                if nodeFound == True:
                    break
        
        listLen = len(unExSet)
        nodeFound = False
        #check if the left position exists and is best
        if (currNode.xPos - 1) > 0:
            #check if node exists in unExSet. If it exists, update its distance value and check if it's the lowest value move
            for i in range(0, (listLen)):
                if unExSet[i].xPos == (currNode.xPos - 1) and unExSet[i].yPos == (currNode.yPos):
                    nodeFound = True
                    unExSet[i].distance = unExSet[i].data
                    if unExSet[i].distance < minVal:
                        minVal = unExSet[i].distance
                        tempNode = unExSet[i] 
                if nodeFound == True:
                    break
        
        listLen = len(unExSet)
        nodeFound = False
        #check if the up position exists and is best
        if (currNode.yPos + 1) < GRAPH_Y_SIZE:
            #check if node exists in unExSet. If it exists, update its distance value and check if it's the lowest value move
            for i in range(0, (listLen)):
                if unExSet[i].yPos == (currNode.yPos + 1) and unExSet[i].xPos == (currNode.xPos):
                    nodeFound = True
                    unExSet[i].distance = unExSet[i].data
                    
                    #prioritizes moving into the final sqare
                    if unExSet[i].xPos == GRAPH_X_SIZE-1 and unExSet[i].yPos == GRAPH_Y_SIZE:
                        minVal = -1
                        tempNode = unExSet[i]
                        
                    if unExSet[i].distance < minVal:
                        minVal = unExSet[i].distance
                        tempNode = unExSet[i]
                    
                if nodeFound == True:
                    break
                    
        #nowhere forward to go - backtrack            
        if minVal == MAX_INT:
            #print("backing up")
            sptSet.pop()
            tempLen = len(sptSet)
            currNode = sptSet[tempLen-1]
            currNode.child = None
        
        else:
            listLen = len(unExSet)
            
            if tempNode.xPos == (GRAPH_X_SIZE-1) and tempNode.yPos == (GRAPH_Y_SIZE-1):
                mazeEnd = True
                finalNode = Node(tempNode.data, tempNode.xPos, tempNode.yPos)
                finalNode.distance = finalNode.data
                sptSet.append(finalNode)
                currNode.child = finalNode
                
                #print("End Found!")
                break
            
            for i in range(0, (listLen-1)):
                if unExSet[i].yPos == tempNode.yPos and unExSet[i].xPos == tempNode.xPos: #and unExSet[i].distance == tempNode.distance:
                    #print(unExSet[i].xPos, unExSet[i].yPos)
                    sptSet.append(unExSet[i])
                    tempLen = len(sptSet)
                    currNode.child = sptSet[tempLen-1]
                    currNode = sptSet[tempLen-1]
                    currNode.parent = sptSet[tempLen-2]
                    unExSet.remove(unExSet[i])
            
        #print(currNode.distance)
        #print("coordinates: ", currNode.xPos, currNode.yPos)
        
        #showing the queue progress after each update to the queue. Showing the X position - Y position - Distance score
        #for THAT SPACE
        print("In Progress Pathway:")
        checkNode = rootNode
        while checkNode.child != None:
            print(checkNode.xPos, " - ", checkNode.yPos, " - ", checkNode.distance)
            checkNode = checkNode.child
        print(checkNode.xPos, " - ", checkNode.yPos, " - ", checkNode.distance)
        print("~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~")
        print()


    
    #print out the final path
    print("The Final Pathway: ")
    print("X - Y - DistVal")
    
    checkNode = rootNode
    totalWeight = 0
    while checkNode.child != None:
        print(checkNode.xPos, " - ", checkNode.yPos, " - ", checkNode.distance)
        totalWeight = totalWeight + checkNode.distance
        checkNode = checkNode.child
    print(checkNode.xPos, " - ", checkNode.yPos, " - ", checkNode.distance)
    totalWeight = totalWeight + checkNode.distance
    print("Total Score of Pathway: ", totalWeight)
```
